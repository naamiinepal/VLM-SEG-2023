# Pandental x-ray images

```python
import os.path
import cv2
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
```

```python
BASE_DIR = "/mnt/Enterprise/PUBLIC_DATASETS/Pandental"
```

```python
img_dir = os.path.join(BASE_DIR, 'Images')
seg1_dir = os.path.join(BASE_DIR, 'Segmentation1')
seg2_dir = os.path.join(BASE_DIR, 'Segmentation2')
```

```python
imgfiles = glob(os.path.join(img_dir, '*.png'))
imgfiles.sort()
seg1files = glob(os.path.join(seg1_dir, '*.png'))
seg1files.sort()
seg2files = glob(os.path.join(seg2_dir, '*.png'))
seg2files.sort()
```

```python
idx = 100
```

```python
img = cv2.imread(imgfiles[idx])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = np.array(img, dtype=np.uint8)
```

```python
seg1 = cv2.imread(seg1files[idx])
seg1 = cv2.cvtColor(seg1, cv2.COLOR_BGR2RGB)
seg1 = np.array(seg1, dtype=np.uint8)
seg1[seg1 > 0] = 255
```

```python
seg2 = cv2.imread(seg2files[idx])
seg2 = cv2.cvtColor(seg2, cv2.COLOR_BGR2RGB)
seg2 = np.array(seg2, dtype=np.uint8)
seg2[seg2 > 0] = 255
```

```python
# plot the images, and their segmentations
fig, ax = plt.subplots(1, 3, figsize=(15, 5), sharey=True, sharex=True)
ax[0].imshow(img)
ax[0].set_title('Image')
ax[1].imshow(seg1)
ax[1].set_title('Segmentation 1')
ax[2].imshow(seg2)
ax[2].set_title('Segmentation 2')
plt.show()
```
